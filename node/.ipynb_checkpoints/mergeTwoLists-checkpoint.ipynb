## 合并两个有序链表

将两个有序链表合并为一个新的有序链表并返回。新链表是通过拼接给定的两个链表的所有节点组成的。 

示例：
```
输入：1->2->4, 1->3->4
输出：1->1->2->3->4->4
```

````python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        h = ListNode(-1)
        cur = h
        
        cur1 = l1
        cur2 = l2
        
        while cur1 is not None and cur2 is not None:
            if cur1.val >= cur2.val:
                cur.next = cur2
                cur2 = cur2.next
            else:
                cur.next = cur1
                cur1 = cur1.next
            cur = cur.next 
        
        if cur1 != None:
            cur.next = cur1
        if cur2 != None:
            cur.next = cur2
        
        return h.next
````

````python
执行用时为 24 ms 的范例
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """

        origin = ListNode(0)
        cur = origin
        while l2 is not None:
            while l1 is not None and l2.val > l1.val:
                cur.next = l1
                cur = cur.next
                l1 = l1.next
            cur.next = l2
            cur = cur.next
            l2 = l2.next
        cur.next = l1 if l2 is None else l2
        return origin.next
````

````python
执行用时为 28 ms 的范例
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        if l1 == None and l2 == None:
            return None
        elif l1 == None:
            return l2
        elif l2 == None:
            return l1
        
        l1_node = l1; l2_node = l2
        rlt_Head = ListNode(0)
        rlt_node = rlt_Head
        
        while l1_node != None or l2_node != None:
            if l1_node == None:
                rlt_node.next = ListNode(l2_node.val)
                l2_node = l2_node.next
            elif l2_node == None:
                rlt_node.next = ListNode(l1_node.val)
                l1_node = l1_node.next
            else:
                if l1_node.val >= l2_node.val:
                    rlt_node.next = ListNode(l2_node.val)
                    l2_node = l2_node.next
                else:
                    rlt_node.next = ListNode(l1_node.val)
                    l1_node = l1_node.next
            rlt_node = rlt_node.next
        return rlt_Head.next
                
````
